1. 데이터 수집 및 이해 (Data Collection & Understanding)
비즈니스 이해: 문제 정의, 목표 설정
데이터 수집: 다양한 소스에서 데이터 수집
데이터 형태 파악: 파일 형식, 크기, 구조 확인
메타데이터 분석: 컬럼 의미, 데이터 타입 이해
2. 탐색적 데이터 분석 (EDA - Exploratory Data Analysis)
2.1 기본 정보 파악
데이터 크기 (행/열 수)
데이터 타입 분포
메모리 사용량 분석
전체 데이터 개요
2.2 단변량 분석 (Univariate Analysis)
수치형 변수:
기술통계량 (평균, 중위수, 분산, 사분위수)
분포 시각화 (히스토그램, 박스플롯)
이상치 탐지
범주형 변수:
빈도수 분석
카테고리 분포 시각화
2.3 결측치 분석 (Missing Value Analysis)
결측치 패턴 분석
결측치 비율 계산
결측치 시각화 (heatmap, bar chart)
결측치 발생 원인 파악
2.4 이변량/다변량 분석 (Bivariate/Multivariate Analysis)
상관관계 분석: 피어슨/스피어만 상관계수
타겟 변수와의 관계: 각 피처와 타겟의 관계
교차분석: 범주형 변수 간 관계
시각화: 산점도, 상관관계 히트맵
3. 데이터 품질 평가 (Data Quality Assessment)
정확성: 데이터 값의 정확도
완전성: 결측치, 누락된 레코드
일관성: 데이터 형식의 일관성
유효성: 비즈니스 룰에 맞는지 확인
중복성: 중복 레코드 탐지
4. 시간 분석 (Temporal Analysis)
시계열 패턴: 트렌드, 계절성, 주기성
시간대별 분석: 시간, 요일, 월별 패턴
이벤트 영향 분석: 특정 시점의 변화
5. 사용자 행동 분석 (User Behavior Analysis)
사용자 세그멘테이션: 행동 패턴별 그룹화
여정 분석: 사용자 행동 흐름
코호트 분석: 시간에 따른 사용자 그룹 변화
6. 비즈니스 인사이트 도출
KPI 분석: 핵심 지표 계산 및 모니터링
A/B 테스트 결과 분석
비즈니스 임팩트 측정
7. 특성 공학을 위한 분석 (Feature Engineering Analysis)
특성 중요도 분석: 어떤 변수가 중요한지
특성 상호작용: 변수 간 상호작용 효과
파생 변수 아이디어: 새로운 변수 생성 가능성
8. 모델링 전 분석 (Pre-modeling Analysis)
타겟 불균형: 클래스 불균형 정도
데이터 분할: Train/Validation/Test 전략
교차검증 전략 수립
9. 고급 분석 기법
9.1 차원 축소 분석
PCA, t-SNE, UMAP을 통한 데이터 구조 파악
9.2 클러스터링 분석
자연스러운 데이터 그룹 발견
9.3 이상치 분석
통계적/머신러닝 기반 이상치 탐지
10. 시각화 및 보고서 작성
대시보드 구성: 핵심 지표 시각화
스토리텔링: 데이터로 이야기 구성
액션 아이템 도출: 분석 결과 기반 실행 계획
🛠️ CTR 예측 프로젝트 특화 분석
CTR 특화 분석 항목들:
클릭률 분포 분석: 전체 CTR, 세그먼트별 CTR
임프레션 분석: 노출 패턴, 빈도 분석
광고/콘텐츠 분석: 카테고리별, 위치별 성능
사용자 세션 분석: 세션 길이, 행동 패턴
시간 기반 분석: 시간대, 요일별 클릭 패턴
디바이스/채널 분석: 플랫폼별 성능 차이
순서 효과 분석: 시퀀스 내 위치별 클릭률
🔧 분석 도구별 활용
Python 생태계:
Pandas: 기본 데이터 조작 및 분석
NumPy: 수치 계산
Matplotlib/Seaborn: 시각화
Plotly: 인터랙티브 시각화
Scipy: 통계 분석
Big Data 도구:
PyArrow: 메모리 효율적 분석
DuckDB: SQL 기반 고속 분석
Polars: 빠른 DataFrame 처리
Spark: 대용량 분산 처리
분석 플랫폼:
Jupyter: 탐색적 분석
Streamlit/Dash: 대시보드
Tableau/PowerBI: 비즈니스 인텔리전스
